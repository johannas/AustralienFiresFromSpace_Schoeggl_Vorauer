```python
# importing necassary libaries
import pandas as pd
from Function_Definitions import *
```

```python
# loading in the rain-, wind- and lightning csv files from the Data\Raw\ directory
rain = pd.read_csv(r'..\Data\Raw\rain.csv')
wind = pd.read_csv(r'..\Data\Raw\wind.csv')
lightning = pd.read_csv(r'..\Data\Raw\lightning.csv')
```

```python
# combining the DF's to a list to loop over them
dfs = [rain, wind, lightning]
```

```python
# renaming Latitude and Longitude into latitude and longitude
for i in dfs: i.rename(columns={'Latitude': 'latitude', 'Longitude': 'longitude'}, inplace=True)
```

```python
# changing the Date/Time column to a datetime object
for i in dfs: i['Date/Time'] = pd.to_datetime(i['Date/Time'])
```

```python
#  splitting the date and time values into separate columns
for i in dfs:
    i['acq_date'] = [j.date() for j in i['Date/Time']]
    i['acq_time'] = [j.time() for j in i['Date/Time']]
```

```python
# dropping unecassary columns
for i in dfs: i.drop(['Number of injuries', 'Number of deaths', 'Insurance payout', 'Damage total', 'Date/Time'], 1, inplace=True)
```

```python
# dropping unique columns
rain.drop(['Rain ID', 'Unnamed: 16'], 1, inplace=True)
wind.drop(['Wind ID', 'Unnamed: 20'], 1, inplace=True)
lightning.drop(['Lightning ID', 'Unnamed: 15'], 1, inplace=True)
```

```python
# adding Year and Month column to the DF's; year_month defined in the Function_Definitions file
for i in dfs: i = year_month(i)
```

```python
# adding season column to the DF's
for i in dfs: i['season'] = ((i.Month+3)%12+3)//3
```
